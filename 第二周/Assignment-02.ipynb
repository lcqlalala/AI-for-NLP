````python
import requests
import re
````

````python
url = "http://www.bjsubway.com/station/zjgls/"
````

````python
#解决爬取的网页乱码问题
hea={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.152 Safari/537.36'}
response = requests.get(url,headers=hea)
response.encoding='gbk'
````

````python
we_want = r'<th>(\w+)——(\w+)</th>'
````

````python
pattern = re.compile(we_want)
````

````python
likes = re.findall(pattern,response.text)
print(type(likes))
````

````python
for like in likes:
    print(like)
````

````python
station_set = set()
for like in likes:
    like_1,like_2 = like
    station_set.add(like_1)
    station_set.add(like_2)
````

````python
from collections import defaultdict
road_dict = defaultdict(list)
for station in station_set:
    for like in likes:
        if station == like[0]:
            road_dict[station] = [like[1]]
        if station == like[1]:
            road_dict[station] = [like[0]]
        else:continue
    
    for like in likes:
        if station == like[1] and like[0] not in road_dict[station]:
            road_dict[station].append(like[0])
        if station == like[0] and like[1] not in road_dict[station]:
            road_dict[station].append(like[1])        
        else:continue
````

````python
road_dict
````

````python
%matplotlib inline
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei'] 
````

````python
import networkx as nx
station_with_road = nx.Graph(road_dict)
nx.draw(station_with_road,with_labels=True,node_size=30)
````

````python
def search(start, destination, connection_grpah, sort_candidate):
    pathes = [[start]]
    
    visited = set()
    
    while pathes:
        path = pathes.pop(0)
        froninter = path[-1]
        
        if froninter in visited: continue
        
        successors = connection_grpah[froninter]
        
        for city in successors:
            if city in visited: continue
            
            new_path = path + [city]
            pathes.append(new_path)
            
            if city == destination: return new_path
        visited.add(froninter)
        pathes = sort_candidate(pathes)
````

````python
def transfer_stations_first(pathes): 
    return sorted(pathes, key=len)
````

````python
search('北京西站','褡裢坡',road_dict,sort_candidate=transfer_stations_first)
````
